## Importing Libraries

```python
import idx2numpy
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

### Ignoring undesired warning

```python
from warnings import simplefilter
simplefilter(action='ignore')
```

### Loading the dataset using 'idx numpy' package

```python
images = idx2numpy.convert_from_file('Datasets/MNIST/images.idx3-ubyte')
label = idx2numpy.convert_from_file('Datasets/MNIST/labels.idx1-ubyte')
print('The size of dataset is : ', images.shape)
```

### Visualise 2 images from the dataset

```python
def img_plot(img):
    plt.figure()
    plt.imshow(img)
    plt.show()


img_plot(images[0])
img_plot(images[1])
```

### Choosing subset of whole dataset to only work with

```python
images_subset = images[0:10000]
label_subset = label[0:10000]
```

### Adjusting image pixels into one

```python
x = []
y = label_subset
for i in range(len(images_subset)):
    x.append(images_subset[i].flatten())

print("The final shape of dataset to work with: ( {0}, {1})".format(len(x), len(x[0])))
```

### TSNE

```python
tsne = TSNE(n_components=2).fit_transform(x)
data = {
    'x': tsne[:, 0],
    'y': tsne[:, 1],
    'digit': y
}

# Plotting
sns.scatterplot(x='x', y='y', hue='digit', legend='full', data=data)
plt.title('Scatter plot of MNIST DATA (on subset of 10,000)')
plt.legend(bbox_to_anchor=(1.01, 1), borderaxespad=0)
plt.show()
```
